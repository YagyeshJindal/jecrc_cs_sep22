```python
import matplotlib.pyplot as plt
import numpy as np
```

# types of graphs
1. plot          :  line graph , between 2 parameter
2. bar           :  bar graph of single parameter  (optional - categorical data)
3. hist          :  histogram of a bunch of data i.e. frequency of single parameter
4. scatter       :  dot graph between 2 parameters 
5. boxplot/box   :  statistical overview in form of box for a single parameter [mean , outliers , quantities (25% , 75%), IQR] (optional - categorical data)
6. stem          :  more like a bar graph , lines instead of bars
7. violinplot    :  violin shaped statistical overview graph , spread of data (more like hist) , single parameter (optional - categorical data)

```python
x = np.linspace(0,5,10)
x1 = np.arange(1,11,3)
```

```python
x
```

```python
x1
```

```python
y = x**3
```

```python
y
```

```python
plt.plot(x,y)
# plt.show() [when not in jupyter]
```

```python
plt.bar(x,y)
plt.title('bar graph')
plt.grid()
plt.xlabel('x -->')
plt.ylabel('y -->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.hist(y)
plt.title('hist')
plt.grid()
plt.xlabel('x -->')
plt.ylabel('y -->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.scatter(x,y)
plt.title('Scatter')
plt.grid()
plt.xlabel('x -->')
plt.ylabel('y -->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.boxplot(x)
```

```python
plt.stem(x,y)
```

```python
plt.violinplot(x)
```

```python
y2 = x**4
```

```python
plt.plot(x,y,x,y2)
plt.legend(['y = x**3', 'y = x**4'])
plt.show()
```

```python
fig = plt.figure(figsize = (10,5))
plt.plot(x,y, lw=3,  marker = 'o' , markersize = 4 , markerfacecolor = 'purple')
plt.plot(x,y2, marker = 'v' , markersize = 2)
plt.legend(['y = x**3','y = x**4'])
plt.grid()
plt.show()
```
