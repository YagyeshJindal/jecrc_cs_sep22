# concat

```python
people1 = {'id' : [1,2,3,4],
           'name' : ['ab','rf','kl','ed'],
           'age' : [45,25,41,30]
          }
people2 = {'id' : [11,22,33,44],
           'name' : ['awb','rgf','kal','ejd'],
           'age' : [35,15,40,35]
          }
people3 = {'id' : [111,222,333,444],
           'name' : ['cwb','rnf','kml','ecd'],
           'age' : [55,75,81,50]
          }
```

```python
df1 = pd.DataFrame(people1)
df2 = pd.DataFrame(people2)
df3 = pd.DataFrame(people3)
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],ignore_index=True)
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
people4 = {'id' : [1,2,3,4],
           'name' : ['ab','rf','kl','ed'],
           'age' : [45,25,41,30]
          }
people5 = {'id' : [1,2,3,4],
           'name' : ['ab','rf','kl','ed'],
           'sales' : [35,15,40,35]
          }
```

```python
df4 = pd.DataFrame(people4)
df5 = pd.DataFrame(people5)
```

```python
pd.merge(df4,df5,on=['id','name'])
```

```python
people6 = {'di' : [1,2,3,4],
           'nema' : ['ab','rf','kl','ed'],
           'age' : [45,25,41,30]
          }
people7 = {'id' : [11,22,33,44],
           'name' : ['awb','rgf','kal','ejd'],
           'sales' : [35,15,40,35]
          }
```

```python
df6 = pd.DataFrame(people6)
df7 = pd.DataFrame(people7)
```

```python
df6.join(df7)
```
